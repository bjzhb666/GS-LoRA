### 探究数据集切分

```python
import torch
from torchvision.datasets import ImageFolder
from torchvision.transforms import ToTensor

def split_dataset(dataset, class_order_list, split1_start, split1_end, split2_start, split2_end):
    # 获取原始数据集的类别数量
    num_classes = len(dataset.classes)

    # 切分区间1的类别索引范围
    split1_class_indices = class_order_list[split1_start:split1_end+1]

    # 创建切分区间1的数据集
    split1_samples = [(sample, label) for sample, label in dataset.samples if label in split1_class_indices]
    split1_dataset = ImageFolder(root=dataset.root, transform=ToTensor())
    split1_dataset.samples = split1_samples
    split1_dataset.targets = [label for _, label in split1_samples]
    split1_dataset.classes = [dataset.classes[idx] for idx in split1_class_indices]
    split1_dataset.class_to_idx = {class_name: i for i, class_name in enumerate(split1_dataset.classes)}

    # 切分区间2的类别索引范围
    split2_class_indices = class_order_list[split2_start:split2_end+1]

    # 创建切分区间2的数据集
    split2_samples = [(sample, label) for sample, label in dataset.samples if label in split2_class_indices]
    split2_dataset = ImageFolder(root=dataset.root, transform=ToTensor())
    split2_dataset.samples = split2_samples
    split2_dataset.targets = [label for _, label in split2_samples]
    split2_dataset.classes = [dataset.classes[idx] for idx in split2_class_indices]
    split2_dataset.class_to_idx = {class_name: i for i, class_name in enumerate(split2_dataset.classes)}

    return split1_dataset, split2_dataset
```

```python
# 假设你已经定义了一个名为 "dataset" 的 ImageFolder 数据集对象
class_order_list = list(range(100))  # 类别顺序列表，从0到99
split1_start = 20  # 区间1起始位置
split1_end = 79  # 区间1结束位置
split2_start = 80  # 区间2起始位置
split2_end = 99  # 区间2结束位置
import torchvision.datasets as datasets
import torchvision.transforms as transforms
data_transform = transforms.Compose([
        transforms.ToTensor(),
    ])
dataset= datasets.ImageFolder(root='./data/faces_webface_112x112_sub100/', transform=data_transform)
split1_dataset, split2_dataset = split_dataset(dataset, class_order_list, split1_start, split1_end, split2_start, split2_end)
```

切分数据集成功

```python
import os
import shutil
import random

# 设置文件夹路径和切分比例
path_to_folder = '/data/zhaohongbo/Github/amnesic-face-recognition/Face-Transformer/data/faces_webface_112x112_sub100'
split_ratio = 0.8  # 训练集占总体的比例

# 创建train_test文件夹
os.makedirs('/data/zhaohongbo/Github/amnesic-face-recognition/Face-Transformer/data/faces_webface_112x112_sub100_train_test', exist_ok=True)
train_test_folder = '/data/zhaohongbo/Github/amnesic-face-recognition/Face-Transformer/data/faces_webface_112x112_sub100_train_test/'

# 创建train和test文件夹
os.makedirs(train_test_folder + 'train', exist_ok=True)
os.makedirs(train_test_folder + 'test', exist_ok=True)
train_folder = train_test_folder + 'train/'
test_folder = train_test_folder + 'test/'

# 遍历原始文件夹下的子文件夹
for subdir in os.listdir(path_to_folder):
    subdir_path = os.path.join(path_to_folder, subdir)
    if os.path.isdir(subdir_path):
        # 在train和test文件夹下创建相应的子文件夹
        os.makedirs(train_folder + subdir, exist_ok=True)
        os.makedirs(test_folder + subdir, exist_ok=True)

        # 获取子文件夹中的所有图片文件
        images = [f for f in os.listdir(subdir_path) if f.endswith('.jpg') or f.endswith('.png')]

        # 计算切分数量
        split_index = int(len(images) * split_ratio)

        # 随机打乱图片顺序
        random.shuffle(images)

        # 将图片移动到train和test文件夹
        for i, image in enumerate(images):
            src_path = os.path.join(subdir_path, image)
            if i < split_index:
                dst_path = os.path.join(train_folder + subdir, image)
            else:
                dst_path = os.path.join(test_folder + subdir, image)
            shutil.copy(src_path, dst_path)

print('训练集和测试集切分完成。')
```

```python
import os
import filecmp

# 两个文件夹的路径
folder1_path = './data/faces_webface_112x112_sub100_train_test/train/'
folder2_path = './data/faces_webface_112x112_sub100_train_test/test/'

# 获取两个文件夹的子文件夹列表
subfolders1 = [f for f in os.listdir(folder1_path) if os.path.isdir(os.path.join(folder1_path, f))]
subfolders2 = [f for f in os.listdir(folder2_path) if os.path.isdir(os.path.join(folder2_path, f))]

# 遍历相同的子文件夹
common_subfolders = set(subfolders1) & set(subfolders2)
assert common_subfolders==set(subfolders1), '两个文件夹中的子文件夹不相同！'

for subfolder in common_subfolders:
    subfolder_path1 = os.path.join(folder1_path, subfolder)
    subfolder_path2 = os.path.join(folder2_path, subfolder)

    # 获取子文件夹中的文件列表
    files1 = [f for f in os.listdir(subfolder_path1) if os.path.isfile(os.path.join(subfolder_path1, f))]
    files2 = [f for f in os.listdir(subfolder_path2) if os.path.isfile(os.path.join(subfolder_path2, f))]

    # 比较子文件夹中的文件是否相同
    diff_files = filecmp.cmpfiles(subfolder_path1, subfolder_path2, files1)

    # 如果有相同的文件存在，则输出结果
    if diff_files[0]:
        print(f"子文件夹 '{subfolder}' 中存在相同的文件：")
        for file in diff_files[0]:
            print(file)
```
