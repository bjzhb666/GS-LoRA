生成自己的子数据集

```python
import os
import shutil

# 定义父文件夹路径
parent_folder = "/data/zhaohongbo/Github/amnesic-face-recognition/Face-Transformer/data/faces_webface_112x112/imgs"

# 获取父文件夹下的所有子文件夹
subfolders = [f.path for f in os.scandir(parent_folder) if f.is_dir()]

# 创建一个字典用于保存每个子文件夹中的样本数量
sample_counts = {}

# 统计每个子文件夹中的样本数量
for folder in subfolders:
    class_name = os.path.basename(folder)
    files = [f for f in os.listdir(folder) if os.path.isfile(os.path.join(folder, f))]
    sample_counts[class_name] = len(files)

# 根据样本数量进行排序，选出样本数量最多的10个类别
sorted_classes = sorted(sample_counts, key=sample_counts.get, reverse=True)
top_10_classes = sorted_classes[:100]

# 创建一个新的文件夹来保存子数据集
output_folder = "/data/zhaohongbo/Github/amnesic-face-recognition/Face-Transformer/data/faces_webface_112x112_sub100"
os.makedirs(output_folder, exist_ok=True)

# 复制样本数量最多的10个类别的子文件夹到输出文件夹中
for class_name in top_10_classes:
    source_folder = os.path.join(parent_folder, class_name)
    destination_folder = os.path.join(output_folder, class_name)
    shutil.copytree(source_folder, destination_folder)

print("子数据集保存成功！")
```
